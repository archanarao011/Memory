Memory games are those which require players to match similar elements. 

Each person turns over two tiles at a time, with the goal of turning over a matching pair, by using their memory.

Rules of the game :

Turn over any two tiles.
If the two tiles match, they remain face up.
If not, they are flipped face down.
The game is over when all the tiles have been matched.
Enter "exit" or "quit" to stop the program.

This game is designed for a single player and can be played in a jupyter notebook or from console.

The actual memory tile GUI is represented as a 4x4 matrix where all the positions will be displayed by default as an asterisk symbol when the game loads intially.




# Programming the game 

There are 2 core functions used in the program.

tileselection() -  for validating the user inputs , and handling the logic for matching the tiles.

displayboard() - to handle the display i.e to retain a face up of matched tiles and face down of wrongly revealed ones.


Throughout the game , the user has to enter values in terms of coordinates ranging from 00 , 01 , 02 .........31, 32, 33.

The tile will turn over to display alphabets in the range of A to H.


If any input other than a 2-digit integer is entered , it prompts the user to enter a valid choice.

An 'exit' or 'quit' will take you out of the program.



Incase you happen to enter the same value for both tiles, the game will instruct you accordingly.



If a match is not found ,the board clears off the revealed unmatched tiles. 
Once a matching pair is found , they remain faced up during the rest of the program. 



The game will be over only when the user matches all the tiles in the board unless he intentionally exits the program.

The time taken to finish will be displayed in seconds. And the user will be asked if he wishes to play again or not.
